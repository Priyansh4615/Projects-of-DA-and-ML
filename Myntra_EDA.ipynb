## 3. Data Analysis - Numeric Column

#### **3.1 Average Rating of a product(non-zero)**

```python
df['rating'].mean()
```

```python
rat =0
count =0

for i in df['rating']:

  if( i !=0):
    rat += i
    count += 1

print(round(int(rat) / count , 2))
```

```python
int(rat)
```

```python
# total items

count
```

```python
# to get the average rating since there are many zero rating , using .mean() will not be a good idea.
# So we have summed up all the non zero rating and have also counted the number of items
# Now, doing Total rating / total items will give us average rating

int(rat) / count
```

```python
# no. of products where rating is zero

len(df) - count
```

### 3.8 Finding the brands with most numbers of products

```python
c = 0

# finding no. of products of each brand


for i in df['brand_name']:
  if (i == 'Greenfibre'):
    c += 1
  
print(" Greenfibre :", c)
```

```python
data = []

for brand_name in df['brand_name'].unique():

  c = 0
  for i in df['brand_name']:
    if ( i == brand_name):
      c += 1

      # print(brand_name, c)

  data.append([brand_name , c])

data = pd.DataFrame(data, columns = ['name' , 'products'])
data.head(10) 
```

```python
data.sort_values(by = 'products', ascending = False).head(10)
```

```python
import numpy as np

pincode = []
state   = []


for i in df['seller_address']:

  if(i != 'Not Mentioned'):
    pincode.append(i.split(' ')[-1])
    state.append(i.split(',')[-1].split('-')[0].strip())

  else:

    pincode.append(np.nan)
    state.append(np.nan)

df['pincode'] = pincode
df['state']   = state 
```
